## 2. EDA-Exploratory Data Analysis

EDA (Exploratory Data Analysis) is a process on data to discover patterns, detect anomalies, test hypotheses and test assumptions. Although it is a task that can be cross-cutting, it is essential prior to the modeling task. EDA is primarily used to see what the data can reveal. It provides a better understanding of the variables in the data set and the relationships between them. It can help you determine if the modeling techniques or analysis method you are considering is appropriate. Helps define data cleaning, transformation and selection steps. It confirms or defines the modeling processes. Understand if the proposed objectives are achievable. Discover errors or limitations of the data. Clarify and delimit the problem (What question are you trying to solve?). Helps answer: How can you add, transform, select or remove features to get more out of your data?

The EDA process includes:

### Convert Unix timestamp to more readable GTM format

```python
list_gtm_dates = ['first_hit_time_gmt', 'hit_time_gmt', 'last_hit_time_gmt', 'post_cust_hit_time_gmt']
```

```python
import pandas as pd
from datetime import datetime

df_log_result['date_time'] = pd.to_datetime(df_log_result['date_time'])

for col_gtm in list_gtm_dates:    
    
    # Formatting gtm Date
    df_log_result[col_gtm  +  '_datetime'] = pd.to_datetime(df_log_result[col_gtm], unit='s', utc=True)    
    df_log_result[col_gtm  +  '_formatted'] = df_log_result[col_gtm  +  '_datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')        
    
    # Converting to datetime
    df_log_result[col_gtm  +  '_datetime'] = pd.to_datetime(df_log_result[col_gtm  +  '_datetime'])
    df_log_result[col_gtm  +  '_formatted'] = pd.to_datetime(df_log_result[col_gtm  +  '_formatted'])

    # Extracting date, houre and minute
    df_log_result[col_gtm  +  '_formatted' + '_date'] = df_log_result[col_gtm  +  '_formatted'].dt.date        
    df_log_result[col_gtm  +  '_formatted' + '_date'] = pd.to_datetime(df_log_result[col_gtm  +  '_formatted' + '_date'])
    df_log_result[col_gtm  +  '_formatted' + '_hour'] = df_log_result[col_gtm  +  '_formatted'].dt.hour        
    df_log_result[col_gtm  +  '_formatted' + '_minute'] = df_log_result[col_gtm  +  '_formatted'].dt.minute      
```

### Total views / hits Month, week and day

```python
# Calculate total visits per day
visits_per_day = df_log_result.groupby(df_log_result['date_time'].dt.date).size()

# Calculate total visits per week
visits_per_week = df_log_result.groupby(df_log_result['date_time'].dt.isocalendar().week).size()

# Calculate total visits per month
visits_per_month = df_log_result.groupby(df_log_result['date_time'].dt.month).size()

# Print the results
print('Visits per day:\n', visits_per_day, '\n', '-'*50)
print('Visits per week:\n', visits_per_week, '\n', '-'*50)
print('Visits per month:\n', visits_per_month, '\n', '-'*50)
```

```python
import matplotlib.pyplot as plt

# Configura un área de gráficos con 1 fila y 3 columnas
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Gráfico de Visitas por Día
visits_per_day.plot(kind='line', ax=axes[0])
axes[0].set_title('Total Visits per Day')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Number of Visits')
axes[0].grid(True)

# Gráfico de Visitas por Semana
visits_per_week.plot(kind='line', ax=axes[1])
axes[1].set_title('Total Visits per Week')
axes[1].set_xlabel('Week of the Year')
axes[1].set_ylabel('Number of Visits')
axes[1].grid(True)

# Gráfico de Visitas por Mes
visits_per_month.plot(kind='line', ax=axes[2])
axes[2].set_title('Total Visits per Month')
axes[2].set_xlabel('Month')
axes[2].set_ylabel('Number of Visits')
axes[2].grid(True)

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```

### Average session duration

```python
# Order datetimes
df_log_result.sort_values('date_time', inplace=True)
df_log_result['diff_seconds'] = df_log_result['date_time'].diff().dt.total_seconds()
df_log_result['diff_seconds'].head()

# Set max time between event  (e.i., 30 minutos) each session.
SESSION_TIMEOUT = 30 * 60  # 30 minutes in seconds

# Cualquier diferencia de tiempo mayor al tiempo máximo se considerará el inicio de una nueva sesión
df_log_result['session_diff'] = df_log_result['diff_seconds'].apply(lambda x: x if x < SESSION_TIMEOUT else None)

# Add the time differences to obtain the total duration of each session.
# Here I assume that each change of day is a new session.
df_log_result['session_id'] = df_log_result['date_time'].dt.date

session_durations = df_log_result.groupby('session_id')['session_diff'].sum()

# Calculate the average session duration
average_session_duration_seconds = session_durations.mean()
average_session_duration_minutes = average_session_duration_seconds / 60  # Convert seconds to minutes

print('average_session_duration_minutes:', average_session_duration_minutes)
```

It can be said that the average time of duration per user session is 18 minutes, which is good for a website that offers medical and health services.

Innovative strategies could be introduced within the platform to make the stay more attractive and not only offer basic and necessary functionalities to offer services, but also additional services such as interesting applications that make predictions of life expectancy of users, according to their medical history, diet, eating habits, among others.

## Dim Time

Is an auxiliary table to describe in more detail the data and make a better analysis of the data over time. For example to evaluate user behavior on weekends (is_weekend), starts/ends of the month (is_start_of_month, is_end_of_month) and holidays (is_holiday).

```python
import pandas as pd
from pandas.tseries.holiday import USFederalHolidayCalendar as Calendar

# Get min date
end_date = df_log_result['first_hit_time_gmt_formatted_date'].max()

# Get max date
start_date = df_log_result['first_hit_time_gmt_formatted_date'].min()

# Create a daily date range between start and end dates
date_range = pd.date_range(start=start_date, end=end_date)

# Create a DataFrame from the date range
df_time_dimension = pd.DataFrame(date_range, columns=['date'])

# Add time-related columns
df_time_dimension['year'] = df_time_dimension['date'].dt.year
df_time_dimension['month'] = df_time_dimension['date'].dt.month_name()
df_time_dimension['day_of_week'] = df_time_dimension['date'].dt.day_name()
df_time_dimension['day_of_month'] = df_time_dimension['date'].dt.day
df_time_dimension['day_of_year'] = df_time_dimension['date'].dt.dayofyear
df_time_dimension['week_of_year'] = df_time_dimension['date'].dt.isocalendar().week
df_time_dimension['quarter'] = df_time_dimension['date'].dt.quarter
df_time_dimension['is_weekend'] = df_time_dimension['date'].dt.weekday > 4
df_time_dimension['is_start_of_month'] = df_time_dimension['date'].dt.is_month_start
df_time_dimension['is_end_of_month'] = df_time_dimension['date'].dt.is_month_end


# US Federal Holidays Calendar
cal = Calendar()
holidays = cal.holidays(start=df_time_dimension['date'].min(), end=df_time_dimension['date'].max())
df_time_dimension['is_holiday'] = df_time_dimension['date'].isin(holidays)

# Display the first few rows of the DataFrame
df_time_dimension.head()
```
